```python
from sklearn.preprocessing import LabelEncoder
le_option=LabelEncoder()
le_training=LabelEncoder()
df['PlayerPosition']=le_option.fit_transform(df['PlayerPosition'])
df['TrainingStatus']=le_training.fit_transform(df['TrainingStatus'])
df
```

```python
df.drop('StrengthBW_multiple',axis=1,inplace=True)
```

```python
a=df.iloc[:,:-1].values
x=df.iloc[:,:-1]
y=df.iloc[:,-1].values
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(a,y,random_state=1,test_size=0.30)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from imblearn.over_sampling import ADASYN
sampling=ADASYN(random_state=42)
x_train,y_train=sampling.fit_resample(x_train,y_train)
```

```python
# from sklearn.ensemble import RandomForestClassifier
# model=RandomForestClassifier(criterion='entropy')
# model.fit(x_train,y_train)
# y_pred=model.predict(x_test)
# y_pred
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
from sklearn.naive_bayes import MultinomialNB
naive=MultinomialNB()
from sklearn.svm import SVC
model1=SVC()
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier(criterion='entropy')
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier(criterion='entropy')
from sklearn.metrics import accuracy_score,confusion_matrix
lst=[knn,naive,model1,model2,model3]
lst1=[]
for i in lst:
    print('algorithm used : ',i)
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    cm=confusion_matrix(y_test,y_pred)
    print('Confusion Matrix')
    print(cm)
    score=accuracy_score(y_test,y_pred)
    print('accuracy score : ',score)
    lst1.append(score)
    print('*****************************************************************************')
```

```python
ax=sns.barplot(x=['knn','naive bayes','svm','tree','forest'],y=lst1,color='red')
plt.xlabel('Algorithms')
plt.ylabel('Accuracy Score')
plt.bar_label(ax.containers[0],fmt='%.2f')
```
